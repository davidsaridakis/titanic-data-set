```python
# Data Handling
import numpy as np
import pandas as pd

# Visualization
import matplotlib as plt
import seaborn as sns
```

```python
# Load and explore the data, check for missing values and interesting 
# patterns for various columns in the data.


# Load CSV files
train_path = "data/raw/train.csv"
test_path = "data/raw/test.csv"
gender_submission_path = "data/raw/gender_submission.csv"

train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)


# Check loaded files.
print("Train shape: ", train_data.shape)
print("Test shape: ", test_data.shape)
train_data.head()

# -------------Explore data----------------

# Check the assumption of the gender_submission, that all females survived.
women = train_data.loc[train_data['Sex'] == 'female', 'Survived']
rate_women = sum(women) / len(women)

print("% of women who survived:", rate_women)

# Check what percentage of men survived.
men = train_data.loc[train_data['Sex'] == 'male', 'Survived']
rate_men = sum(men) / len(men)

print ("% of men who survived:", rate_men)

# Graph survival rate by sex
#sns.barplot(x='Sex', y='Survived', data=train)

#  Check survival based on passenger class.
train_data.groupby('Pclass')["Survived"].mean()

# Check survival by (binned) age.
train_data.groupby(pd.cut(train_data["Age"], bins=5))['Survived'].mean()

# Survival based on combined effect of sex and passenger class.
train_data.groupby(['Sex', 'Pclass'])['Survived'].mean()

# View the summary stats.
train_data.describe()

# Check for missing values in the data.
train_data.isnull().sum()






```
